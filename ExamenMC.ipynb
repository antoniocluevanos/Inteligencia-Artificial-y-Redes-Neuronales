```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv')
```

```python
print(df.head(5))
```

```python
print(df.dtypes)
```

```python
print(df['horsepower'].isnull().sum())
```

```python
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
```

```python
df['relacion_peso_potencia'] = df['weight'] / df['horsepower']
print(df.head(5))
```

```python
print(df.isnull().sum())
```

*Parte 2*

```python
plt.hist(df['acceleration'],bins=30,color='darkgreen')
plt.title("Distribución de Aceleración")
plt.xlabel("Aceleración")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.scatter(df['displacement'], df['horsepower'],alpha=0.9,color='lightgreen')
plt.title("Desplazamiento vs Caballos de Fuerza")
plt.xlabel("Desplazamiento")
plt.ylabel("Caballos de Fuerza")
plt.show()
```

```python
promedio_mpg = df.groupby('origin')['mpg'].mean()

promedio_mpg.plot(kind='bar',color='darkgreen')
plt.title("Promedio MPG por Origen")
plt.xlabel("Origen")
plt.ylabel("MPG Promedio")

plt.show()
```
